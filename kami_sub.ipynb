## 【問題1】べき乗の算術演算子を使用した実装

```python
# 紙を43回折った時の厚さを計算するコード

THICKNESS = 0.00008

folded_thickness = THICKNESS * 2**43

print("厚さ： {}メートル".format(folded_thickness))
```

## 【問題2】単位の変換

```python
# メートルをキロメートルに変換して小数点以下2桁で表示する
print("厚さ： {:.2f}万キロメートル".format(folded_thickness / 1000 / 10000))
```

## 【問題3】for文を使用して作成

```python
accum = THICKNESS
for _ in range(43):
    accum *= 2
folded_thickness = accum

print("厚さ： {:.2f}万キロメートル".format(folded_thickness / 1000 / 10000))
```

## 【問題4】計算時間の比較

```python
import time
```

```python
start = time.time()

THICKNESS * 2**43

elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```python
start = time.time()

accum = THICKNESS
for _ in range(43):
    accum *= 2

elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

差が出ないので反復数を増やす

```python
start = time.time()

for _ in range(10000):
    THICKNESS * 2**43

elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```python
start = time.time()

for _ in range(10000):
    accum = THICKNESS
    for _ in range(43):
        accum *= 2

elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

## 【問題5】リストへの保存

```python
accum = THICKNESS
ys = [accum]
for _ in range(43):
    accum *= 2
    ys.append(accum)
print(ys)
len(ys)
```

## 【問題6】折れ線グラフの表示

```python
import matplotlib.pyplot as plt
%matplotlib inline

# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(ys)  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

一回折るごとに倍々になるため、40 回より前ではほとんど厚さが変化していないように見える

## 【問題7】グラフのカスタマイズ

```python
# 水色の折れ線グラフを表示する。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(ys, color='cyan')
plt.show()
```

```python
# 緑の線で対数軸にする
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.yscale('log')
plt.plot(ys, color='green')
plt.show()
```

```python
# ラベルを大きくする
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(labelsize=14)
plt.yscale('log')
plt.plot(ys, color='green')
plt.show()
```
